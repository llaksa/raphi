```javascript
var TARGET=[1,1,0];
//var ALPHABET = "ABCDEFGHIJKLMNOPQRSTUVWXYZ ";
var MUT_PROB = 10;

var generateGenome = function(){
    var genome = [];
    for (var i =0; i<9;++i){
         genome[i]=Math.random();
    }
    
    return genome;
};  

var genome = generateGenome();

var sensors2img = function(genome){
    // fit1 fit2 y fit3 de la IMG
    // FALTA MODIFICAR!!!
    // JALAR EL ARRAY QUE LE CORRESPONDE A LA IMG RESPECTO DEL GENOME(ARRAY DE SENSORES) INGRESADO!!!
    return [Math.random(), Math.random(), Math.random()]
}

var getFitness = function(genome){
    var fitness=0;
    var fitFactors = sensors2img(genome)
    for (var i=0; i<TARGET.length;++i){
        if (TARGET[i]-fitFactors[i] < 0.2){
            fitness++;
        }
    }
    
    return fitness;
}; 

//getFitness(genome);

var getGenePool = function(genome){
    var pool = [];
    for (var i=0; i<50;++i){
        pool[i] = genome;
    }
    return pool;
};

var pool = getGenePool(genome);

var getFittest = function(pool){
    var fittestLoc = 0;
    var fittest = 0;
    for (var i=0; i<pool.length;++i){
        if (getFitness(pool[i]) >fittest){
            fittest = getFitness(pool[i]);
            fittestLoc = i;
        }
    }
    return pool[fittestLoc];
};

//getFittest(pool);

var doMutation = function(genome){
    var testGenome = genome
    var newGenome = [];
    for (var i=0; i<genome.length;++i){
        if(Math.floor(Math.random()*MUT_PROB)===1){
            testGenome[i]=Math.random();
            if(getFitness(testGenome)>getFitness(genome)){
                newGenome[i]=testGenome[i];
            }
            else{
                newGenome[i]=genome[i];
            }
        } 
        else{
            newGenome[i]=genome[i];
        }
    }
    return newGenome;
};

//doMutation(genome);

var evolve = function(){
    var numGens = 0;
    var fittest = generateGenome();
    console.log('AAAA');
    console.log(fittest);
    
    while(getFitness(fittest)!==3){
        numGens++;
        var pool = getGenePool(fittest);
        var pool2 = [];
        for (var i=0; i<pool.length;++i){
            pool2[i]=doMutation(pool[i]);
        }
        fittest = getFittest(pool2);
        if(numGens%10===0){ 
            console.log('BBBBB');
            console.log(fittest);
        }
    }
    return fittest;
};

evolve();
```
